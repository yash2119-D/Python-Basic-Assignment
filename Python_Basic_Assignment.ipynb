        In essence, the Python interpreter is the engine that runs your Python programs.

11. What is the purpose of conditional statements in Python ?

  - Common Conditional Statements in Python:
  - if statement:

  - Example: If a number is positive, print "Positive".
  - elif statement:
      Stands for "else if" and allows checking multiple conditions if the initial if condition is False.

  - Control Flow: They allow different paths of execution depending on conditions, making the program flexible and interactive.

      In summary, conditional statements allow Python programs to adapt to different situations by evaluating conditions and executing appropriate code accordingly.

12. How does the elif statement work ?
  - The elif statement in Python is used to check multiple conditions after an initial if statement. If the condition in the if statement is False, the elif statement provides an alternative condition to check. It allows for testing more than one condition in sequence, avoiding the need for nested if statements.

  - How it works:
      The if statement checks the first condition. If it's True, its corresponding code block is executed.
      If the if condition is False, Python checks the next elif condition.
      If the elif condition is True, its code block runs.
      If none of the if or elif conditions are True, the else block (if provided) is executed.
  - Key Points:
      You can use multiple elif statements to handle different conditions.
      Only the first True condition will trigger its corresponding block of code; others are ignored.
      The else block is optional and executes if no prior conditions are met.


13.What is the difference between for and while loops ?
  - The  for  loop and the while loop in Python are both used to repeat a block of code, but they differ in how the loop conditions are structured and when they are used.

--- Key Differences:

1.  Condition Type :
  - for loop : Typically used when the number of iterations is known in advance or when iterating over a sequence (e.g., a list, tuple, or range).
  - while  loop : Runs as long as a specified condition is `True`, and the number of iterations is not necessarily known beforehand.

2.  Use Case :
  - for loop : Ideal for iterating over a collection of items or when you know exactly how many times to loop.
  - while` loop : Useful when you want to repeat something until a certain condition changes, without knowing the exact number of iterations.

3.  Structure :
   	-- for  loop : It automatically iterates over a sequence or range of values.
   	-- while  loop : The condition is explicitly checked before each iteration, and it continues looping as long as the condition remains `True`.

 - Example:
  - for  loop : Used to loop over a fixed range or collection.
  		Example: Iterate through a list of numbers.
  
  - while  loop : Used when the loop depends on a condition being met.
        Example: Loop until a variable reaches a specific value.

  - Summary:
      `for` loop :  Iterates over a known sequence or range.
 	    `while` loop : Continues until a specific condition is no longer met.



14.  Describe a scenario where a while loop is more suitable than a for loop.
  - A  while loop  is more suitable than a  for loop  when you need to repeat an action until a certain condition is met, but you don't know in advance how many iterations are required. It’s ideal for situations where the stopping condition is dynamic or based on some event that happens during the loop.

  - Scenario : Waiting for User Input

      Imagine you're creating a program that repeatedly asks the user to enter a valid password. You don't know how many attempts it will take for the user to enter the correct password, so a `while` loop is a better choice.

  - Use Case : The program keeps asking for input until the user provides the correct password.
      In this case, you don't know beforehand how many times the user will need to try, so a `while` loop is ideal because it keeps running as long as the condition (`entered password is not correct`) remains true.

  - Why `while` loop works :
      The number of iterations depends on the user’s input and can vary, so the condition (whether the password is correct or not) controls the loop, rather than iterating a fixed number of times as with a `for` loop.

```python
#Q - 6. Write a function to demonstrate the behavior of mutable and immutable arguments.

def demonstrate_arguments(mutable_arg, immutable_arg):
    """
    Function to show how mutable and immutable arguments behave.
    """
    # Modify the mutable argument (list)
    mutable_arg.append(100)
    print("Inside function - Mutable argument after modification:", mutable_arg)

    # Attempt to modify the immutable argument (integer)
    immutable_arg += 50
    print("Inside function - Immutable argument after modification:", immutable_arg)


# Create a mutable object (list) and an immutable object (integer)
mutable_obj = [1, 2, 3]
immutable_obj = 10

print("Before function call:")
print("Mutable object:", mutable_obj)
print("Immutable object:", immutable_obj)

# Call the function
demonstrate_arguments(mutable_obj, immutable_obj)

print("\nAfter function call:")
print("Mutable object:", mutable_obj)  # Modified outside the function
print("Immutable object:", immutable_obj)  # Unchanged outside the function
```

```python
#Q - 9. Write a Python program to convert user input from string to integer, float, and boolean types.

def convert_input():
    # Getting input from the user as a string
    user_input = input("Enter something: ")

    # Convert string to integer
    try:
        int_value = int(user_input)
        print(f"Converted to integer: {int_value}")
    except ValueError:
        print("Could not convert to integer.")

    # Convert string to float
    try:
        float_value = float(user_input)
        print(f"Converted to float: {float_value}")
    except ValueError:
        print("Could not convert to float.")

    # Convert string to boolean
    # In Python, "True" or "False" (case-sensitive) are the valid string representations of boolean values
    if user_input.lower() == "true":
        bool_value = True
    elif user_input.lower() == "false":
        bool_value = False
    else:
        bool_value = bool(user_input)  # This converts non-empty string to True, empty string to False
    print(f"Converted to boolean: {bool_value}")

# Call the function
convert_input()

```

```python
#Q - 10. Write code to demonstrate type casting with list elements.

# Original list with different data types
mixed_list = ["123", "45.67", 89, True, None]

# Type casting examples
# Convert string to integer
int_list = [int(x) if isinstance(x, str) and x.isdigit() else x for x in mixed_list]
print(f"String to Integer: {int_list}")

# Convert string to float
float_list = [float(x) if isinstance(x, str) else x for x in mixed_list]
print(f"String to Float: {float_list}")

# Convert integer to string
str_list = [str(x) if isinstance(x, int) else x for x in mixed_list]
print(f"Integer to String: {str_list}")

# Convert boolean to integer (True -> 1, False -> 0)
bool_to_int_list = [int(x) if isinstance(x, bool) else x for x in mixed_list]
print(f"Boolean to Integer: {bool_to_int_list}")

# Convert None to string
none_to_str_list = [str(x) if x is None else x for x in mixed_list]
print(f"None to String: {none_to_str_list}")
```

```python
#Q - 13. Write a Python program to find the sum of all even numbers between 1 and 50.
  # Initialize the sum variable
sum_even = 0

# Iterate through numbers from 1 to 50
for num in range(1, 51):
    # Check if the number is even
    if num % 2 == 0:
        sum_even += num  # Add the even number to the sum

# Print the result
print("The sum of all even numbers between 1 and 50 is:", sum_even)
```

```python
#Q - 14. Write a program to reverse a string using a while loop.
  # Function to reverse a string using a while loop
def reverse_string(input_string):
    reversed_string = ""  # Start with an empty string
    index = len(input_string) - 1  # Start from the last character

    # Iterate backwards through the string using a while loop
    while index >= 0:
        reversed_string += input_string[index]  # Append the character to the reversed string
        index -= 1  # Move to the previous character

    return reversed_string

# Example usage
input_string = input("Enter a string: ")
reversed_string = reverse_string(input_string)
print("Reversed string:", reversed_string)
```

```python
#Q - 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
  # Function to calculate factorial using a while loop
def factorial(number):
    result = 1  # Initialize result to 1 (as factorial of 0 or 1 is 1)

    # Check if the number is negative, as factorial is not defined for negative numbers
    if number < 0:
        return "Factorial is not defined for negative numbers"

    # While loop to calculate factorial
    while number > 1:
        result *= number  # Multiply the result by the current number
        number -= 1  # Decrease the number by 1

    return result

# Get input from the user
try:
    num = int(input("Enter a number: "))
    fact = factorial(num)
    print(f"The factorial of {num} is: {fact}")
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```
